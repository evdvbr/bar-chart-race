```python
import pandas as pd
```

```python
# Read csv and store in a dataframe
olympics_dataset_df = pd.read_csv('https://raw.githubusercontent.com/mudit111/bar-chart-race/master/athlete_events.csv')
```

```python
# Display the dataframe
olympics_dataset_df
```

```python
# Filter wrestling data
wrestling_df = olympics_dataset_df[olympics_dataset_df['Sport']=='Wrestling'][olympics_dataset_df.Medal.isin(['Gold','Silver','Bronze'])]
```

```python
# Display the dataframe
wrestling_df
```

```python
# Filter the dataframe columns to select the relevant columns
wrestling_df = wrestling_df[['Team','Year','Medal']]
```

```python
# Display the dataframe
wrestling_df
```

```python
# Pivot dataframe
wrestling_olympics = wrestling_df.groupby(['Team','Year']).count().reset_index().pivot(index='Team',columns='Year',values='Medal').fillna(0).astype(int).reset_index()

# Looping through the columns and adding the previous column to each column
previous_year = wrestling_olympics.columns[1]
for year in wrestling_olympics.columns[2:]:
    wrestling_olympics[year] = wrestling_olympics[previous_year] + wrestling_olympics[year]
    previous_year = year
```

```python
# Display final dataframe
wrestling_olympics
```
